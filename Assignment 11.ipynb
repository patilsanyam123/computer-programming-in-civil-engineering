```python
import random

class AirPollutionControlSystem:
    def _init_(self, pollutant_emission, control_efficiency):
        self.pollutant_emission = pollutant_emission
        self.control_efficiency = control_efficiency

    def apply_control(self):
        control_removal = self.pollutant_emission * self.control_efficiency
        self.pollutant_emission -= control_removal
        return control_removal

class AirQualityMonitor:
    def _init_(self):
        self.pollutant_concentration = 0

    def measure_pollutant_concentration(self):
        self.pollutant_concentration = random.uniform(0, 100)

def main():
    pollutant_emission = random.uniform(50, 100)
    control_efficiency = random.uniform(0.1, 0.9)

    control_system = AirPollutionControlSystem(pollutant_emission, control_efficiency)
    air_monitor = AirQualityMonitor()

    for _ in range(10):
        control_removal = control_system.apply_control()
        air_monitor.measure_pollutant_concentration()

        print(f"Pollutant Emission: {control_system.pollutant_emission:.2f}")
        print(f"Control Efficiency: {control_system.control_efficiency:.2f}")
        print(f"Control Removal: {control_removal:.2f}")
        print(f"Current Pollutant Concentration: {air_monitor.pollutant_concentration:.2f}\n")

if _name_ == "_main_":
    main()
```
